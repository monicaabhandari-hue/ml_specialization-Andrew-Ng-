**Goal:** Learning to implement the model $f_w,_b$ for linear regression with one variable.

**Problem Statement:** We are trying to predict the housing price using a simple data set with only two data points - a house with 1000 square feet sold for \\$300,000 and a house with 2000 square feet sold for \\$500,000. Our goal is to fit a linear regression model through these two points, and then predict price for other houses(a house with 1200 square feet).

## Notation

| Symbol | Description |
|-----------------|-------------|
| **x** | input variable feature values |
| **y** | output variable |
|$x^{(i)}$, $y^{(i)}$ | $i_{th}$Training Example |
| **m** | Number of training examples |
| **w** | Model parameter: weight |
| **b** | Model parameter: bias |
| $f_{w,b}(x^{(i)})$ | The result of the model evaluation at $x^{(i)}$ parameterized by $w,b$: $f_{w,b}(x^{(i)}) = wx^{(i)}+b$  |

```python
# loading the libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_train = np.array([1.0,2.0])
y_train = np.array([300.0,500.0])
print(f"x_train = {x_train}")
print(f"y_train = {y_train}")
```

```python
print(f"x_train.shape: {x_train.shape}")
m = x_train.shape[0]
print(f"Number of training examples is: {m}")
```

```python
# Example: syntax to access location of x_train

i = 0
x_i = x_train[i]
y_i = y_train[i]
print(f"(x^({i}), y^({i})) = ({x_i}, {y_i})")
```

```python
# Plotting the data

plt.scatter(x_train, y_train, marker = 'x', c = 'r')
plt.title("Housing Prices")
plt.ylabel('Price(in 1000s of $)')
plt.xlabel('Size(1000 sqft)')
plt.show
```

**Model Function:** The model function for linear regression is represented as $f_w,_b$(x) = wx+b.

```python
# can adjust the model's w and b parameters

w = 100
b = 100
print(f"w: {w}")
print(f"b: {b}")
```

Computing the value of $f_w,_b(x^{(i)})$ for the two data points. Since, we have only two data points we can explicitly write them. If there are a large number of data points, we calculate the function output in a for loop.

```python
def compute_model_output(x, w, b):
  ''' (ndarray(m,)) describes a NumPy n-dimensional array of shape (m,)
  (scalar), describes an argument without dimensions, just a magnitude. w,b is a scalar.'''

  m = x.shape[0]
  f_wb = np.zeros(m) # np.zeros(m) will return a one-dimensional numpy array with m entries

  for i in range(m):
    f_wb[i] = w * x[i] +b
  return f_wb
```

```python
tmp_f_wb = compute_model_output(x_train, w, b)
plt.plot(x_train, tmp_f_wb, c = 'b', label = 'Our Prediction')
plt.scatter(x_train, y_train, marker = 'x', c = 'r', label = 'Actual Values')
plt.title("Housing Prices")
plt.ylabel('Price (in 1000s of $)')
plt.xlabel('Size (1000 sqft)')
plt.legend()
plt.show()
```

Setting w = 100 and b = 100 does not result in a line that fits our data.
